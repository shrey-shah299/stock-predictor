```python
import pandas as pd
import numpy as np
import matplotlib as pt
import yfinance as yf
import os
```

```python
#downloading data from yfinance two methods
#1.
rasch=yf.Ticker("RCF.NS")
rasch=rasch.history(period="max")
rasch.to_csv("Rashtriya_chemical.csv")
```

```python
rasch.head()
```

```python
rasch.index=pd.to_datetime(rasch.index)
```

```python
rasch
```

```python
rasch.index=rasch.index.date
```

```python
rasch
```

```python
rasch=rasch.drop(columns=["Dividends","Stock Splits"],errors='ignore')
```

```python
rasch
```

```python
rasch.plot.line(y="Close",use_index=True)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
train=rasch.iloc[:-100]
test=rasch.iloc[-100:]
features=["Open","High","Low","Volume"]
target="Close"

lm.fit(train[features],train[target])
```

```python
preds=lm.predict(test[features])
preds=pd.Series(preds,index=test.index)
preds.plot.line(use_index=True)
```

```python
combined=pd.concat([preds,test[target]],axis=1)
combined.plot.line(ylabel="Closing price",use_index=True ,color=["green","red"],label=["Predicted","Actual"])
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

mse=mean_squared_error(test[target],preds)
mae=mean_absolute_error(test[target],preds)
rsq=r2_score(test[target],preds)

print("MSE :",mse)
print("MAE :",mae)
print("rsq :",rsq)
```
